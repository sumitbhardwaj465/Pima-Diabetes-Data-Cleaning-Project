```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("diabetes.csv")
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
cols = ['Glucose','BloodPressure',
        'SkinThickness','Insulin','BMI']

for col in cols:
    print(col, (df[col]==0).sum())
```

```python
sns.countplot(x='Outcome', data=df)
plt.title("Diabetes Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

```python
df.hist(figsize=(12,10))
plt.show()
```

```python
sns.boxplot(x='Outcome',
            y='Glucose',
            data=df)

plt.show()
```

```python
sns.boxplot(x='Outcome',
            y='BMI',
            data=df)

plt.show()
```

```python
sns.boxplot(x='Outcome',
            y='Age',
            data=df)

plt.show()
```

```python
sns.pairplot(df,
             hue='Outcome',
             vars=['Glucose',
                   'BMI',
                   'Age',
                   'Insulin'])

plt.show()
```
